# Visualization Technique (25%)
In this assignment, I will demonstrate how to analyze time series data using **Line and Scatter subplots**, which can give us a general idea of how the data points look over time and whether if there is evidence of trend over time. ## Diving into More Complex Analysis

The earlier two exploration plots only showed basic plotting capabilities of Plotly. We uncovered general idea of question counts per month, but that distribution leaves out much more valuable time-series data. To uncover deeper insights with time, such as understanding trends of question counts over time for each StackOverflow topic, we need more sophisticated plots.

### Prepare Helper Functions

```python
def remove_entries(dictionary : dict):
     '''
        Remove go.Figure axes attributes that mess up subplots. 
     '''
     if "anchor" in dictionary:
          del dictionary["anchor"] 
     
     if "title" in dictionary:
          del dictionary["title"]

     if "domain" in dictionary:
          del dictionary["domain"]

def generate_subplots(figures : list[go.Figure], custom_specs=None, desired_rows=None, desired_columns=None, titles=None, x_title=None, y_title=None) -> go.Figure:

    '''
        For each existing figure, append their traces to new subplot figure
        based on the correct row and col indices
    '''
    num_figures = len(figures)
    # num_traces = sum([len(figure.data) for figure in figures])
    
    # if desired_rows is not None:
    #     desired_columns = num_figures // desired_rows + 1 

    # if desired_columns is not None:
    #     desired_rows = num_figures // desired_columns + 1


    fig = make_subplots(desired_rows, desired_columns,
                        specs=np.full((desired_rows, desired_columns), {}).tolist() if custom_specs is None else custom_specs,
                        subplot_titles=titles,
                        x_title=x_title,
                        y_title=y_title)
    
    row = 1
    col = 1
    for figure in figures:
        if col > desired_columns:
                col = 1
                row += 1
        
        # Add figure to subplot
        for trace in figure["data"]:
            fig.append_trace(
                trace, row=row, col=col
            )
        
        # Remove x-axis and y-axis figure attributes that are not needed
        # and mess up subplots function
        xaxis_dict = figure["layout"]["xaxis"].to_plotly_json()
        remove_entries(xaxis_dict)

        yaxis_dict = figure["layout"]["yaxis"].to_plotly_json()
        remove_entries(yaxis_dict)

        fig.update_xaxes(xaxis_dict, row=row, col=col)
        fig.update_yaxes(yaxis_dict, row=row, col=col)

        col += 1

    # Use desired template for subplots
    fig.update_layout(template="plotly_white")

    return fig

def generate_subplot_topic_question_counts(data : pd.DataFrame, transformation=lambda x: x, x_title=None,  y_title=None) -> go.Figure:
    unique_topics = data["topic"].sort_values(key=lambda x: x.str.lower()).unique()
    figures = []
    figures_dict = {}
    titles = []

    # Generate plots of question count per month over time for each topic
    for i, topic in enumerate(unique_topics):
        topic_counts = data.loc[data["topic"] == topic, ["month", "question_count_per_month"]]
        topic_counts["question_count_per_month"] = transformation(topic_counts["question_count_per_month"])
        fig = px.scatter(topic_counts, x="month", y="question_count_per_month")
        fig.update_traces(mode="lines")
        fig.update_xaxes(tickformat="%b %Y", title=x_title)
        fig.update_yaxes(title=y_title)
        fig.update_layout(title=topic)
        
        figures.append(fig)
        figures_dict[topic] = fig
        titles.append(f"{i}: {topic}")
    
#     print(len(figures))
    return generate_subplots(figures, desired_rows=30, desired_columns=3, titles=titles, x_title=x_title, y_title=y_title), figures_dict
```

## Scatter plot of trend patterns

For each topic, aggregate and calculate average question count every 6 months. We perform min-max normalization on the numbers and calculate the difference between adjacent average question count every 6 months. Let's call the result of this calculation **the pattern score**.

Why does this categorize our time-series data effectively? Because topics showing a clear positive trend in number of questions over time (ex. Pandas) would only increase in average question count every 6 months, and if you take the differences between every adjacent question count, the pattern score will be 1 or close to 1 due to the min-max normalization.

If a topic has a trend that is **more uncertain**, the pattern score will be closer to 0.5. 

If the topic has completely died off in the recent and no one is asking about it on StackOverflow, the pattern score will be closer to 0, because even if your average question count has increased in the past, it completely tapers off towards the end so you have a net zero gain in number of questions asked towards the end.


```python
df_melted.sort_values(by=["topic", "month"], inplace=True)
```

```python
df_melted_6_month_averages = df_melted.set_index("month").groupby("topic").resample("6M")["question_count_per_month"].mean()
```

```python
df_melted_6_month_averages_norm = df_melted_6_month_averages.groupby("topic").transform(lambda x: (x - x.min())/(x.max() - x.min()))
```

```python
df_melted_6_month_averages_diff = df_melted_6_month_averages_norm.groupby("topic").apply(lambda x: pd.DataFrame({"diff_count_per_month" : x - x.shift(1)}))
```

```python
df_melted_6_month_averages_diff = df_melted_6_month_averages_diff.droplevel(1)
```

```python
df_pattern_per_topic = df_melted_6_month_averages_diff.groupby("topic").sum()
```

```python
df_pattern_per_topic.columns = ["pattern_score"]
```

```python
fig = px.scatter(df_pattern_per_topic.reset_index(), x="pattern_score")
fig.update_traces(customdata=df_pattern_per_topic.reset_index()["topic"].to_numpy())
fig.update_traces(hovertemplate="%{customdata}: %{x}")
fig.update_xaxes(title="Pattern Score")
fig.update_yaxes(title="Dummy Index")
fig.update_layout(title="Distribution of Pattern Score")
```

We can see the Pattern Score is definitely meaningful in some way, as there are clear groups formed from the score. Let's add some colours and lines to separate and see the different types of trends visually

```python
def assign_cluster(elem):
    if elem <= 0.2:
        return "Dead Topic"
    elif elem <= 0.5:
        return "Dying Topic"
    elif elem <= 0.8:
        return "Moderately Popular Topic"
    else:
        return "Popular Topic"
    
df_pattern_per_topic["topic_type"] = df_pattern_per_topic["pattern_score"].apply(assign_cluster)
```

```python
df_pattern_per_topic.sort_values(by="topic_type", inplace=True)
```

```python
custom_colors = [px.colors.diverging.RdBu[0], px.colors.diverging.RdBu[2], px.colors.diverging.RdBu[8], px.colors.diverging.RdBu[9]]
```

```python
fig = px.scatter(df_pattern_per_topic.reset_index(), x="pattern_score", color="topic_type", hover_data=["topic"],
                 color_discrete_sequence=custom_colors)
fig.update_traces(hovertemplate="%{customdata}: %{x}") # customdata is my hover data
fig.add_vrect(
    x0=-0.1, x1=0.2,
    line=dict(
        color="white",
        width=2,
    ),
    layer="below",
    annotation_align="left",
    annotation_text="<b>Dead Topics</b>",
    annotation_position="top left",
    annotation_font_color="#dd571c",
    annotation=dict(font_size=20),
    fillcolor="orange",
)
fig.add_vrect(
    x0=0.2, x1=0.5,
    line=dict(
        color="white",
        width=2,
    ),
    layer="below",
    annotation_align="left",
    annotation_text="<b>Dying Topics</b>",
    annotation_position="top left",
    annotation_font_color="#ff601e",
    annotation=dict(font_size=20),
    fillcolor="#FFD580",
)
fig.add_vrect(
    x0=0.5, x1=0.8,
    line=dict(
        color="white",
        width=2,
    ),
    layer="below",
    fillcolor="LightBlue",
    annotation_align="left",
    annotation_text="<b>Moderately Popular</b><br><b>Topics</b>",
    annotation_position="top left",
    annotation_font_color="#448acf",
    annotation=dict(font_size=20)
)
fig.add_vrect(
    x0=0.8, x1=1.03,
    line=dict(
        color="white",
        width=2,
    ),
    layer="below",
    fillcolor="#6bcae2",
    annotation_align="left",
    annotation_text="<b>Popular Topics</b>",
    annotation_position="top left",
    annotation_font_color="#005a9c",
    annotation=dict(font_size=20)
)
fig.update_xaxes(title="Pattern Score")
fig.update_yaxes(title="Dummy Index")   
fig.update_layout(title="Distribution of Pattern Score")
```

Now we have a clearer plot that qualitatively describes trends of StackOverflow topics, with colored rectangles and markers. 

Once again, the plotly charts are very easy to configure. I can easily customize my hovertext, and add coloured groups I simply add a 'color' key and specify which column I want to distinguish.

## Bar Chart plots to compare differences between different types of topics
- Compare average question count percent increases per month
- Compare average question count percent decreases per month

```python
df_melted_index = df_melted.set_index("topic")
compare_average_question_count = pd.merge(df_melted_index, df_pattern_per_topic["topic_type"], left_index=True, right_index=True)
```

```python
df_diff_per_topic_group = df_melted.groupby("topic")["question_count_per_month"].transform(lambda x: np.where(x.shift(1) > 0, (x - x.shift(1))/x.shift(1)* 100, 
                                                                                                              np.where(x - x.shift(0) == 0, 0, 100)))
df_diff_per_topic_group.index = df_melted["topic"]
compare_diff_question_count = pd.merge(df_diff_per_topic_group, df_pattern_per_topic["topic_type"], left_index=True, right_index=True)
```

```python
percent_increases_by_topic_type = compare_diff_question_count.groupby("topic_type")["question_count_per_month"].agg(
    [lambda x: x.loc[x > 0].mean(), lambda x: len(x.index.unique())]
    # [lambda x: x.loc[x > 0].mean(), lambda x: x.loc[x > 0].std)]
)
percent_increases_by_topic_type.columns = ["question_count_average_percent_increase_per_month", "number_of_unique_topics"]
percent_increases_by_topic_type
```

```python
percent_decreases_by_topic_type = compare_diff_question_count.groupby("topic_type")["question_count_per_month"].agg(
    lambda x: x.loc[x < 0].mean()
)
percent_decreases_by_topic_type.name = "question_count_average_percent_decrease_per_month"
```

```python
increase_decrease_df = pd.merge(percent_decreases_by_topic_type, percent_increases_by_topic_type, left_index=True, right_index=True)
```

```python
increase_decrease_df
```

```python
increase_decrease_df_melt = pd.melt(increase_decrease_df.reset_index(), id_vars=["topic_type","number_of_unique_topics"], var_name="metric_type", value_name="Average Value")
```

```python
increase_decrease_df_melt
```

```python
fig = px.bar(increase_decrease_df_melt, x="topic_type", y="Average Value", color="metric_type", hover_data="number_of_unique_topics")
fig.update_yaxes(title="Average Percent Change (%)")
fig.update_layout(barmode="group", title="<b>Average Percent Increase and Decrease</b> of StackOverflow question count<br>per month by Topic Popularity")
```

Notice how "Dead Topics" are characterized by high average increase, meaning topics that are not longer asked on StackOverflow in recent memory generally once had a rapid growing trend of questions. 

On the contrary, "Popular Topics" are characterized by more stable increase of question counts per month

"Moderately Popular Topics" seem to both rapidly increase and decrease in number of questions asked over time,

While "Dying Topics', like "Popular Topics", face stable increase and decrease of question counts per month. Maybe in hindsight, it's not appropriate to call these topics "Dying Topics", since it's still relatively stable. So maybe these are just "Not As Popular Topics".

A majority of topics are "Popular Topics" on StackOverflow (50/81 topics)

# Conclusion

We uncovered many different trends of topics being asked on StackOverflow. We first tried to understand trend of StackOverflow question counts over time for each topic by plotting the time series for every topic, but we quickly found ourselves overwhelmed by the quantity of visualizations.

Thus, we calculated a **pattern score** to uncover different trend patterns per topic more compactly and visualize different types of trends. We were able to understand far more about the types of trends through the **pattern_score** compared to a time series plot for every topic.

Last but not least, we used a bar chart to understand more about the types of trends we identified from the **pattern_score**. We realized that "Dying Topics" was not an accurate name for topics with pattern_score between 0.2 and 0.5, which at first glance is a fitting name, but due to the relative stability of counts per question trend for topics under this umbrella, "Not As Popular Topics" maybe be a better name, since "Dying" suggests the topic is decreasing in question counts per month very quickly.
